```python
import pandas as pd
import numpy as np
import statsmodels.api as sm


df = pd.read_csv(r"C:\Users\ADMIN\Desktop\auto data\Auto.csv")

df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df_clean = df.dropna(subset=['horsepower'])

# 2. Define Target (Y) and Predictors (X)
Y = df_clean['mpg']
# Using multiple numeric predictors
X = df_clean[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin']]

# 3. Add constant for the intercept (Essential for statsmodels)
X = sm.add_constant(X)
```

```python
Model = sm.OLS(Y,X).fit()
```

```python
print(Model.summary())
```
